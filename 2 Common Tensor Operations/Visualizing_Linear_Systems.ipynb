```python
# =========================
# VISUALISASI SISTEM LINEAR
# =========================

import numpy as np              # import numpy untuk perhitungan numerik
import matplotlib.pyplot as plt # import matplotlib untuk visualisasi

# membuat nilai x (range nilai x)
x = np.linspace(-10, 10, 100)  # 100 titik dari -10 sampai 10

# persamaan pertama: y = 2x + 1
y1 = 2*x + 1  # hitung nilai y berdasarkan x

# persamaan kedua: 3x - y = 5 -> y = 3x - 5
y2 = 3*x - 5  # ubah ke bentuk y = mx + c lalu hitung

# membuat plot garis
plt.plot(x, y1, label="y = 2x + 1")   # plot garis pertama
plt.plot(x, y2, label="y = 3x - 5")   # plot garis kedua

# mencari titik potong (opsional)
# 2x + 1 = 3x - 5 → x = 6
x_intersect = 6                    # nilai x titik potong
y_intersect = 2*x_intersect + 1    # substitusi ke salah satu persamaan

# menampilkan titik potong
plt.scatter(x_intersect, y_intersect)  # titik potong
plt.text(x_intersect, y_intersect, f"({x_intersect}, {y_intersect})")  # label titik

# menambahkan elemen visual
plt.axhline(0)  # garis horizontal (sumbu x)
plt.axvline(0)  # garis vertikal (sumbu y)
plt.legend()    # menampilkan legenda
plt.title("Visualisasi Sistem Persamaan Linear")  # judul
plt.xlabel("x")  # label sumbu x
plt.ylabel("y")  # label sumbu y
plt.grid()       # menampilkan grid

# menampilkan grafik
plt.show()
```
