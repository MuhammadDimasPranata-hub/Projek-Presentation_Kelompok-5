```python
import numpy as np              # import numpy untuk perhitungan numerik
import matplotlib.pyplot as plt # import matplotlib untuk visualisasi
```

```python
x = np.linspace(-10, 10, 1000) # start, finish, n points
```

```python
y1 = 2*x + 1
y2 = 3*x - 5
```

```python
fig, ax = plt.subplots()
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([3, 10])
ax.set_ylim([7, 15])
ax.plot(x, y1, c='green')
ax.plot(x, y2, c='brown')
plt.axvline(x=6, color='purple', linestyle='--')
_ = plt.axhline(y=13, color='purple', linestyle='--')
```
